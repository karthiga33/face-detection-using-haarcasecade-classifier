```python
#Importing cv2 Package
import cv2

from google.colab.patches import cv2_imshow

```

```python
#Reading the Image
#Printing the Image


faceCase = cv2.CascadeClassifier('/content/haarcasecade.xml')
img = cv2.imread('/content/img.jpeg')

print(img)
```

```python
#Converting Image to Gray Scale

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

```

```python
# Loading the required haar-cascade xml classifier file

faces = faceCase.detectMultiScale(gray,1.1,4)
i = 0
for [x,y,w,h] in faces :


    cv2.rectangle(img,(x,y),(x+w, y+h),(0,255,0,0),2)

```
